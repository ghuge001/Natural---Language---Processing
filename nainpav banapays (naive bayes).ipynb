```python
def wordcl(data, title):
    txt = ' '.join(data['text'])
    txt = txt.lower()
    stop = STOPWORDS
    wc  = WordCloud(background_color= 'white', stopwords=stop).generate(txt)
    plt.imshow(wc)
    plt.axis('off')
    plt.title(title)
```

```python
wordcl(spam, 'SPAM')
```

```python
wordcl(ham, 'HAM')
```

```python
lemma = WordNetLemmatizer()
```

```python
# Data Cleaning
corpus = []
for i in range (len(df)):
    txt = re.sub('[^a-zA-z]',' ', df['text'][i])
    txt = txt.lower()
    txt = txt.split()
    
    txt = [lemma.lemmatize(i) for i in txt if not i in stopwords.words('english')]
    txt = ' '.join(txt)
    corpus.append(txt)
```

```python
corpus
```

```python
# Model Building
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
cv = CountVectorizer()
x = cv.fit_transform(corpus)

x = pd.DataFrame(x.toarray(), columns=cv.get_feature_names_out())
x
```

```python
y = df['label']
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
nb = MultinomialNB()

nb.fit(xtrain,ytrain)
ypred = nb.predict(xtest)

print(nb.score(xtrain,ytrain))
print(nb.score(xtest,ytest))
print(classification_report(ytest,ypred))
```
